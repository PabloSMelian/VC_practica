```python
vid = cv2.VideoCapture(0)

#Dimensiones de la cÃ¡mara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Fuerzo a mitad de resoluciÃ³n para ocupar menos pantalla
vid.set(cv2.CAP_PROP_FRAME_WIDTH, w) #En Mac no reacciona a estos comandos
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

#Imagen conjunta 2x original
collage = np.zeros((h,w,3), dtype = np.uint8)
tl = collage[0:h,0:w]

while True:      
    # fotograma a fotograma
    ret, frameIN = vid.read()

    #Menor tamaÃ±o
    frame = cv2.resize(frameIN, (int(w),int(h)),cv2.INTER_NEAREST)

    if ret:
        #Separamos canales
        r = frame[:,:,2]
        g = frame[:,:,1]
        b = frame[:,:,0]

        #Jugamos
        tl[:,:,0] = 255-b
        tl[:,:,1] = 255-g
        tl[:,:,2] = 255-r
    
        # Muestra composicion
        cv2.imshow('Cam', collage)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

```python
vid = cv2.VideoCapture(0)

# TipografÃ­a para mostrar texto
font = cv2.FONT_HERSHEY_SIMPLEX
  
lanzado = 0
px = -1
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret: 
        # Activa em manejador en el primer fotograma
        if lanzado == 0:
            # Muestra fotograma
            cv2.imshow('Cam', frame)   
            # Define el nombre del manejador del evento
            cv2.setMouseCallback('Cam', mouse_events)
            lanzado = 1
            
        # Muestra valores RGB intentando centrar en el puntero
        if px > -1:
            cv2.putText(frame, '{}'.format(R), (px-45,py-5), font, 0.5, (0, 0, 255), 1)
            cv2.putText(frame, '     {}'.format(G), (px-54,py-5), font, 0.5, (0, 255, 0), 1)
            cv2.putText(frame, '         {}'.format(B), (px-54,py-5), font, 0.5, (255, 0, 0), 1)
            
        cv2.imshow('Cam', frame)   
        
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

TAREA:
Pintar cÃ­rculos en las posiciones del pÃ­xel mÃ¡s claro y oscuro de la imagen 
Â¿Si quisieras hacerlo sobre la zona 8x8 mÃ¡s clara/oscura?

```python
cap = cv2.VideoCapture(0)  # 0 representa la cÃ¡mara predeterminada

while True:
    # Capturar un fotograma de la cÃ¡mara
    ret, frame = cap.read()

    # Comprueba si la captura fue exitosa
    if not ret:
        break

    # Encuentra la posiciÃ³n del pÃ­xel mÃ¡s oscuro en la imagen
    min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY))
    center_x_dark, center_y_dark = min_loc  # PosiciÃ³n del pÃ­xel mÃ¡s oscuro

    # Dibujar un cÃ­rculo en la posiciÃ³n del pÃ­xel mÃ¡s oscuro
    radius = 10  # Radio del cÃ­rculo
    color = (0, 0, 255)  # Color en formato BGR (Rojo en este caso)
    thickness = 2  # Grosor de la lÃ­nea del cÃ­rculo
    cv2.circle(frame, (center_x_dark, center_y_dark), radius, color, thickness)

    center_x_clear, center_y_clear = max_loc  # PosiciÃ³n del pÃ­xel mÃ¡s oscuro

    # Dibujar un cÃ­rculo en la posiciÃ³n del pÃ­xel mÃ¡s oscuro
    radius = 10  # Radio del cÃ­rculo
    color = (0, 255, 0)  # Color en formato BGR (Verde en este caso)
    thickness = 2  # Grosor de la lÃ­nea del cÃ­rculo
    cv2.circle(frame, (center_x_clear, center_y_clear), radius, color, thickness)

    # Mostrar la imagen con el cÃ­rculo
    cv2.imshow('CÃ¡mara con CÃ­rculo en PÃ­xel MÃ¡s Oscuro y otro en el mas claro', frame)

    # Esperar a que se presione la tecla 'q' para salir
    if cv2.waitKey(20) == 27:
        break
  
# Liberar la cÃ¡mara y cerrar la ventana
cap.release()
cv2.destroyAllWindows()
```

Salida de la webcam "inspirada" en Andy Warhol

```python
vid = cv2.VideoCapture(0)

#Dimensiones de la cÃ¡mara
w = int(vid.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(vid.get(cv2.CAP_PROP_FRAME_HEIGHT))

#Fuerzo a mitad de resoluciÃ³n para ocupar menos pantalla
w=int(w/2)
h=int(h/2)
vid.set(cv2.CAP_PROP_FRAME_WIDTH, w) #En Mac no reacciona a estos comandos
vid.set(cv2.CAP_PROP_FRAME_HEIGHT, h)

#Imagen conjunta 2x original
collage = np.zeros((h*2,w*2,3), dtype = np.uint8)
tl = collage[0:h,0:w]
tr = collage[0:h,w:w+w]
bl = collage[h:h+h,0:w]
br = collage[h:h+h,w:w+w]

while True:      
    # fotograma a fotograma
    ret, frameIN = vid.read()

    #Menor tamaÃ±o
    frame = cv2.resize(frameIN, (int(w),int(h)),cv2.INTER_NEAREST)

    if ret:
        #Separamos canales
        r = frame[:,:,2]
        g = frame[:,:,1]
        b = frame[:,:,0]

        #Jugamos
        tl[:,:,0] = b
        tl[:,:,1] = g
        tl[:,:,2] = r

        tr[:,:,0] = 255 - r
        tr[:,:,1] = g
        tr[:,:,2] = b
        
        bl[:,:,0] = r
        bl[:,:,1] = 255 - b
        bl[:,:,2] = g

        br[:,:,0] = b
        br[:,:,1] = g
        br[:,:,2] = 255 - r
    
        # Muestra composicion
        cv2.imshow('Cam', collage)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```

Transformando la imagen capturada en una colecciÃ³n de cÃ­rculos, cuyo radio depende de la intensidad promedio de la zona

```python
vid = cv2.VideoCapture(0)

ncells = 10
off = int(ncells/2)
  
while(True):      
    # fotograma a fotograma
    ret, frame = vid.read()

    if ret:
        #Dimensiones originales
        h, w, c = frame.shape
        #Redimensiono
        down_frame = cv2.resize(frame, (int(w/ncells),int(h/ncells)),cv2.INTER_NEAREST)
        #Dimensiones reducidas
        h2, w2, c2 = down_frame.shape

        #Separamos canales
        r = down_frame[:,:,0]
        g = down_frame[:,:,1]
        b = down_frame[:,:,2]

        # Creamos imagen negra
        gris_up_frame = np.zeros((h2*ncells,w2*ncells,1), dtype = np.uint8)

        for y in range(0,h2):
            for x in range(0,w2):
                # La suma de los valores RGB define el radio del cÃ­rculo
                rad = int ( (r[y,x] + g[y,x] + b[y,x])/(ncells*3*2))
                cv2.circle(gris_up_frame,(x*ncells+off, y*ncells + off), rad, (255,255,255), -1)

    
        # Muestra fotograma resultante
        cv2.imshow('Cam', gris_up_frame)
    
    # Detenemos pulsado ESC
    if cv2.waitKey(20) == 27:
        break
  
# Libera el objeto de captura
vid.release()
# Destruye ventanas
cv2.destroyAllWindows()
```
